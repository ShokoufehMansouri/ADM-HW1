```python
#Problem 1.
#Basic Data Types
#Find the Runner-Up Score!:
if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())
arr=list(arr)
arr.sort()
p=arr[n-1]
p1=arr[0]
for i in range(1,n):
    if arr[i]<p and arr[i]>p1:
        p1=arr[i] 
print(p1)
```

```python
#Problem 1.
#Basic Data Types
#Nested Lists:
if __name__ == '__main__':
    l=0
    grade=[]
    for _ in range(int(input())):
        l=l+1
        name = input()
        score = float(input())
        grade.append([name,score])
grade.sort(key = lambda x: x[1])
g1=grade[1]
p1=g1[1]
g=grade[0]
p=g[1]
for i in range(2,l):
    y=grade[i]
    if p==p1:
        p1=y[1]
    elif y[1]>p and y[1]<p1:
        p1=y[1] 
u=[]
for i in range(l):
    y=grade[i]
    if y[1]==p1:
        u.append(y[0])
u.sort()
for i in range(len(u)):
    print(u[i])
```

```python
#Problem 1.
#Basic Data Types
#Finding the percentage:
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()

a=sum(student_marks[query_name])
b=len(student_marks[query_name])
average=a/b
print("%.2f" % average)
```

```python
#Problem 1.
#Strings
#sWAP cASE:
def swap_case(s):
    l = list(s) 
    for i in range(len(l)):
        if l[i].isupper():
            l[i]=s[i].lower()
        else:
            l[i]=l[i].upper()
    s = ''.join(l)
    return s

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
#Problem 1.
#Strings
#Mutations:
def mutate_string(string, position, character):
    string = string[:position] + character + string[position+1:]
    return string

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
#Problem 1.
#Strings
#Find a string:
def count_substring(string, sub_string):
    n=0
    while string.find(sub_string)!=-1:
        n=n+1
        string=string[string.find(sub_string)+1:]
    return n

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
#Problem 1.
#Strings
#String Validators:
if __name__ == '__main__':
    s = input()
    p=[False]*5
    for j in range(len(s)):
        if s[j].isalnum():
            p[0]=True
            break
    for j in range(len(s)):
        if s[j].isalpha():
            p[1]=True
            break
    for j in range(len(s)):
        if s[j].isdigit():
            p[2]=True
            break
    for j in range(len(s)):
        if s[j].islower():
            p[3]=True
            break
    for j in range(len(s)):
        if s[j].isupper():
            p[4]=True
            break
    for i in range(5):
        print(p[i])
```

```python
#Problem 1.
#Strings
#Text Alignment:

thickness = int(input()) #This must be an odd number
c = 'H'
#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

```python
#Problem 1.
#Strings
#Text Wrap:
import textwrap

def wrap(string, max_width):
    a=textwrap.wrap(string.strip(),max_width)
    w=a[0]
    for i in range(1,len(a)):
        w = '\n'.join([w, a[i]])
    return w

if __name__ == '__main__':
    string, max_width = input(), int(input())
    result = wrap(string, max_width)
    print(result)
```
